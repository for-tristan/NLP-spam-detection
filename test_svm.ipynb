```python
import pickle
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('stopwords', quiet=True)

# نفس دالة التنظيف بالظبط
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    tokens = text.split()
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words and len(word) > 1]
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(word) for word in tokens]
    return ' '.join(tokens)

# تحميل الملفين اللي عملناهم
with open("svm_model.pkl", "rb") as f:
    model = pickle.load(f)
with open("svm_vectorizer.pkl", "rb") as f:
    vectorizer = pickle.load(f)

def check_spam(message):
    cleaned = preprocess_text(message)
    vectorized = vectorizer.transform([cleaned])
    prediction = model.predict(vectorized)[0]
    if prediction == 1:
        return "🔴 Spam"
    else:
        return "🟢 Not Spam"

print("Model is ready to test! ✅")
```

```python
msg1 = "Congratulations! You've won a FREE $1000 Walmart gift card. Go to http://bit.ly/123 to claim now."
print(f"Message: {msg1}")
print(f"Result:  {check_spam(msg1)}")
```
